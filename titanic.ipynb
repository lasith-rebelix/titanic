```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
data.head()
```

```python
data.info()
```

```python
def clean(data):
    data = data.drop(["Ticket","Cabin","PassengerId"],axis=1)
    cols = ("SibSp","Parch","Fare","Age")
    for col in cols:
        data[col].fillna(data[col].median(),inplace=True)
    data.Embarked.fillna("U",inplace=True)
    return data

data=clean(data)
test=clean(test)
    
    
```

```python
data.head()
```

```python
data.info()
```

```python
from sklearn import preprocessing

# Instantiate LabelEncoder
le = preprocessing.LabelEncoder()

# Columns to encode
cols = ["Sex", "Embarked"]

# Encode categorical variables in both training and testing datasets
for col in cols:
    data[col] = le.fit_transform(data[col])  # Fit and transform training data
    test[col] = le.transform(test[col])      # Transform testing data
    print(le.classes_)                        # Print the classes learned by LabelEncoder

# Display the head of the training dataset
data.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = data.drop(["Survived","Name"], axis=1)
y = data["Survived"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)


```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
